```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib as mpl
from matplotlib import cm
import seaborn as sns
from plotly.validators.bar import marker

sns.set_theme(font_scale=2.0, rc={'text.usetex': True})

plt.rcParams['text.usetex'] = True
plt.rcParams['font.size'] = 20  # Set the font size to 14
plt.rcParams['text.latex.preamble'] = r'\usepackage{lmodern}'
```

```python
x = np.loadtxt('../code/output/coin_flips_count_leading_dist.txt')

sns.set_theme(style='ticks')

f, ax = plt.subplots(figsize=(3, 3))

sns.histplot(
    x,
    bins=np.arange(0, 32, dtype=int),
    color='k',
    alpha=0.5,
    edgecolor='0',
    linewidth=.4,
    log_scale=(False, False),
    stat='density'
)
ax.spines['top'].set_linewidth(0.5)
ax.spines['right'].set_linewidth(0.5)
ax.spines['left'].set_linewidth(0.5)
ax.spines['bottom'].set_linewidth(0.5)
ax.tick_params(width=0.5)

ax.set_title("Coin Flips For Real")
ax.set_xlabel('Number Of Heads Before First Failure')
ax.set_ylabel('Density')
ax.grid(True)
ax.set_xlim(0, 10)
ax.set_ylim(0, 0.55)

plt.savefig(ax.get_title().lower().replace(' ', '_') + '.pdf', bbox_inches='tight')
```

```python
x1 = np.loadtxt('../code/output/coin_flips_count_leading_dist.txt')
x2 = np.loadtxt('../code/output/coin_flips_for_real_dist.txt')

# Create a figure and axis
hist1, bin_edges1 = np.histogram(x1, bins=np.arange(0, 32, dtype=int), density=True)
hist2, bin_edges2 = np.histogram(x2, bins=np.arange(0, 32, dtype=int), density=True)

# Create a figure and axis
f, ax = plt.subplots(figsize=(3, 3))
sns.set_theme(style='ticks')


# Plot the first histogram
sns.lineplot(
    x=bin_edges1[:-1],
    y=hist1,
    linewidth=0.5,
    color='k',
    marker='o',
    markersize=8,
    markeredgewidth=1,
    markerfacecolor='none',
    markeredgecolor='k',
    label='Count Leading Zeros',
    ax=ax
)

# Plot the second histogram
sns.lineplot(
    x=bin_edges2[:-1],
    y=hist2,
    linewidth=0.5,
    color='k',
    marker='s',
    markersize=9,
    markerfacecolor='none',
    markeredgecolor='k',
    markeredgewidth=1,
    label='Coin Flips For Real',
    ax=ax
)

ax.spines['top'].set_linewidth(0.5)
ax.spines['right'].set_linewidth(0.5)
ax.spines['left'].set_linewidth(0.5)
ax.spines['bottom'].set_linewidth(0.5)
ax.tick_params(width=0.5)

ax.set_title("Comparison of Distributions")
ax.set_xlabel('Number of Heads / Leading Zeros')
ax.set_ylabel('Density')
ax.grid(True)
ax.legend()
ax.set_xlim(0, 10)
ax.set_ylim(0, 0.55)

plt.savefig(ax.get_title().lower().replace(' ', '_') + '.pdf', bbox_inches='tight')
```
